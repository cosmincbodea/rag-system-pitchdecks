```python
import fitz  # PyMuPDF
from PIL import Image
import os

# Define the paths
pitchdecks_path = '../data/pitchdecks/'
slides_path = '../data/slides/'

# Ensure the slides directory exists
os.makedirs(slides_path, exist_ok=True)

# Function to convert PDF to images
def pdf_to_images(pdf_path, output_folder, base_name):
    pdf_document = fitz.open(pdf_path)
    for page_num in range(len(pdf_document)):
        page = pdf_document.load_page(page_num)
        pix = page.get_pixmap()
        output_image_path = os.path.join(output_folder, f"{base_name}_slide{page_num + 1}.png")
        pix.save(output_image_path)
    pdf_document.close()

# Process each PDF in the pitchdecks folder
for filename in os.listdir(pitchdecks_path):
    if filename.endswith('.pdf'):
        pdf_path = os.path.join(pitchdecks_path, filename)
        base_name = os.path.splitext(filename)[0]
        pdf_to_images(pdf_path, slides_path, base_name)
```
